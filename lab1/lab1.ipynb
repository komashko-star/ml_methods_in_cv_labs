## Задание №2

> - Возьмите любое изображение из интернета, в котором присутствует хотя бы один объект. Воспользуйтесь функциями из OpenCV для выделения объекта(ов)

```python
img = cv2.imread('cat.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

t = img_rgb.copy()

# кошка
cat_point1 = (500, 1000)
cat_point2 = (2500, 3000)

cv2.rectangle(t, cat_point1, cat_point2, (255, 0, 0), 15)

# кокос
coconut_point1 = (550, 2600)
coconut_point2 = (1800, 3700)

cv2.rectangle(t, coconut_point1, coconut_point2, (255, 0, 0), 15)

plt.figure(figsize=(3, 4))
plt.imshow(t)

plt.show()
```

## Задание №3

> - Воспроизведите функцию rectangle из OpenCV с нуля. Разрешено использовать только NumPy для разработки из сторонних библиотек (не обязательно, чтобы параметры были те же самые, как в OpenCV, например thickness. Главное, чтобы прямоугольник было видно на изображении). Продемонстрируйте работу вашей функции на тестовом изображении

```python
def draw_rectangle(image, top_left, bottom_right, color, thickness=1):
    x1, y1 = top_left
    x2, y2 = bottom_right
    
    h, w = image.shape[:2]
    
    x1, x2 = sorted([x1, x2])
    y1, y2 = sorted([y1, y2])

    for t in range(thickness):
        if 0 <= y1 + t < h:
            image[y1 + t, max(x1, 0):min(x2 + 1, w)] = color
        if 0 <= y2 - t < h:
            image[y2 - t, max(x1, 0):min(x2 + 1, w)] = color

    for t in range(thickness):
        if 0 <= x1 + t < w:
            image[max(y1, 0):min(y2 + 1, h), x1 + t] = color
        if 0 <= x2 - t < w:
            image[max(y1, 0):min(y2 + 1, h), x2 - t] = color

    return image
```

```python
t = img_rgb.copy()
draw_rectangle(t, cat_point1, cat_point2, (0, 0, 255), 30)
draw_rectangle(t, coconut_point1, coconut_point2, (0, 255, 0), 15)
draw_rectangle(t, (2000, 2000), (4000, 5000), (0, 255, 255), 15)

plt.figure(figsize=(3, 4))
plt.imshow(t)

plt.show()
```
